```python
%matplotlib inline

import matplotlib.pyplot as plt

from anchovy.crosssection import SectionArray


station = [0, 0, 300, 300, 600, 600, 900, 900]
elevation = [9, 3, 3, 0, 0, 3, 3, 9]
sect_stat = [300, 600]

array = SectionArray(station, elevation)

for sa in array.split(sect_stat):
    s, e = sa.coordinates()
    _ = plt.plot(s, e)
    _ = plt.plot(s[0], e[0], ls='None', marker='s')
    _ = plt.plot(s[-1], e[-1], ls='None', marker='x', markersize=10)

_ = plt.xlabel('Station')
_ = plt.ylabel('Elevation')
```

```python
from anchovy.crosssection import CrossSection


roughness = [0.07, 0.035, 0.07]

xs = CrossSection(station, elevation, roughness, sect_stat)
_ = xs.plot(6)
```

```python
import matplotlib.pyplot as plt
import numpy as np

rough_elev = np.linspace(0, 9)

_ = plt.plot(rough_elev, xs.roughness(rough_elev))
_ = plt.xlabel('Elevation')
_ = plt.ylabel('Roughness')
```
