```python
%matplotlib inline

import matplotlib.pyplot as plt

from hydxscomp.frict import TableFrict


station = [0, 0, 300, 300, 600, 600, 900, 900]
elevation = [9, 3, 3, 0, 0, 3, 3, 9]
sect_stat = [300, 600]

roughness = [0.035, 0.035, 0.07]
rough_elev = [0, 3, 9]
frict = TableFrict(rough_elev, roughness)
frict.plot()
```

```python
from hydxscomp.crosssection import CrossSection


xs1 = CrossSection(station, elevation, frict)
_ = xs1.plot(6)
```

```python
xs2 = CrossSection(station, elevation, frict, sect_stat)
_ = xs2.plot(6)
```

```python
import numpy as np


stage = np.linspace(0, 9)
k1 = xs1.conveyance(stage)
k2 = xs2.conveyance(stage)

plt.plot(k1, stage, label='Total XS conveyance')
plt.plot(k2, stage, label='Sum of SS conveyance')
plt.xlabel('Conveyance, in cfs')
plt.ylabel('Stage, in ft')
plt.legend()
```
